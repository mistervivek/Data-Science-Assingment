```python
from sklearn import preprocessing

label_encoder=preprocessing.LabelEncoder()
ff['month'] = label_encoder.fit_transform(ff['month'])
ff['day'] = label_encoder.fit_transform(ff['day'])
ff['size_category'] = label_encoder.fit_transform(ff['size_category'])
```

```python
ff
```

```python
ff=ff.drop(ff.columns[0:2],axis=1)
ff.head()
```

```python
ff.info()
```

# Splitting the data into x and y as input and output# Splitting the data into x and y as input and output

```python
x= ff.iloc[:,0:28]
y = ff.iloc[:,28]
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
array_x=scaler.fit_transform(x)
```

```python
array_x
```

```python
x=pd.DataFrame(array_x)
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=0)
```

```python
clf = SVC()
param_grid = [{'kernel':['rbf'],'gamma':[60,50,5,10,0.5],'C':[20,15,14,13,12,11,10,0.1,0.001] }]
gsv = GridSearchCV(clf,param_grid,cv=10)
gsv.fit(x_train,y_train)
```

```python
gsv.best_params_ , gsv.best_score_ 
```

```python
clf = SVC(C= 20, gamma = 60)
clf.fit(x_train , y_train)
y_pred = clf.predict(x_test)
acc = accuracy_score(y_test, y_pred) * 100
print("Accuracy =", acc)
confusion_matrix(y_test, y_pred)
```
